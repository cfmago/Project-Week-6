```python
heatmap_cor = {#'respondent_id' : data['respondent_id'],
               'more_time_hobbies_LD' : data['more_time_hobbies_LD'],
               'more_time_TV_LD' : data['more_time_TV_LD'],
               'more_time_internet_LD' : data['more_time_internet_LD']}
               
heatmap_cor = pd.DataFrame(heatmap_cor)

fig, ax = plt.subplots(figsize=(20,10))
sns.heatmap(heatmap_cor.corr(),annot=True, xticklabels = heatmap_cor.corr().columns,yticklabels = heatmap_cor.corr().columns, cmap = 'YlGnBu')

plt.savefig('Figures/heatmap_moretime.jpg')

plt.show()
```
